## Preguntas objetivo del negocio

## Gráfico 1. Distribución porcentual de fumadores

```python
fumadores = df['fumador'].value_counts()
etiquetas = fumadores.index
colors = sns.color_palette("hls")
plt.title('Distribución porcentual de fumadores')
plt.pie(fumadores, labels = etiquetas, colors = colors, autopct='%.0f%%')
plt.show()
```

La mayor cantidad de clientes no son fumadores

## Gráfico 2. Valor seguro contra fumador

```python
Fumador_seguro = sns.boxplot(x=df["fumador"], y=df["valor_seguro"], palette = sns.color_palette("hls"))
plt.title('Valor seguro contra fumador')
plt.xlabel('Fumador')
plt.ylabel('Valor Seguro')
plt.savefig('boxplot.png')
```

Los fumadores pagan un seguro mucho más costoso

## Gráfico 3. Rangos de edad

```python
total_rango_edad = df['Rango_Edad'].groupby(df['Rango_Edad']).count()
labels = total_rango_edad.index
colors = sns.color_palette('hls')

plt.pie(total_rango_edad, labels = labels, colors = colors, autopct='%.1f%%')
plt.title('Rangos de edad')
plt.savefig('circular.png')
plt.show()
```

El rango de edad A representa cerca del 30 % del total de registros

## Gráfico 4. Total de hijos

```python
total_hijos = df['hijos'].groupby(df['hijos']).count()
etiquetas = total_hijos.index
colors = sns.color_palette('hls')
plt.title('Total de hijos')
plt.pie(total_hijos, labels = etiquetas, colors = colors, autopct='%.0f%%')
plt.show()
```

- Cerca del 45 % de los clientes no tienen hijos

- Aproximadamente, el 70 % de los clientes tiene al menos un hijo.

## Gráfico 5. Valor seguro contra fumador

```python
fuma_seg_pc = df.groupby('fumador')['valor_seguro'].sum()
etiquetas = fuma_seg_pc.index
colors = sns.color_palette("hls")
plt.title('Valor seguro contra fumador')
plt.pie(fuma_seg_pc, labels = etiquetas, colors = colors, autopct='%.1f%%')
plt.show()
```

Aunque hay más personas no fumadoras en el conjunto de datos, el valor del seguro que pagan los fumadores es casi igual al de los no fumadores.
Esto indica que, a pesar de ser menos en cantidad, los fumadores representan un alto potencial económico para la venta de seguros, ya que sus pólizas suelen ser más costosas debido al mayor riesgo asociado al hábito de fumar.

## Gráfico 6. Relación entre variables

```python
sns.heatmap(df.corr(numeric_only = True),annot=True, cmap="BuPu")
plt.title('Mapa de calor: Relación entre variables')
plt.show()
```

La relación más cercana al valor número 1 es la edad y el índice de masa corporal con respecto al valor pagado al seguro

## Gráfico 7. Relación tres variables: valor seguro, rango de edad y el índice de masa corporal

```python
sns.relplot(x='Rango_Edad' ,y='valor_seguro',hue='imc',data=df)
plt.title('Relación de valor seguro, rango de edad e IMC')
```

Los valores más altos pagados corresponden a todos los rangos de edad con un índice de masa corporal alto.

## Gráfico 8. Relación de tres variables: valor seguro, rango de edad y el indicador de fumador

```python
sns.relplot(x='Rango_Edad', y='valor_seguro',hue='fumador',data=df, palette=sns.color_palette("hls"))
plt.title('Relación de valor seguro, rango de edad y fumador')
```

En cualquier rango de edad, el hecho de que sea fumador es determinante para el pago del valor del seguro.

## Gráfico 9. Relación región y estado fumador con los valores del seguro

```python
sns.barplot(data=df,x='region',y='valor_seguro',hue='fumador', palette = sns.color_palette("hls"))
plt.title('Relación región y estado fumador con los valores del seguro')
```

El hábito de fumar incrementa notablemente el costo del seguro, sin importar la región.

## Gráfico 10. Histogramas variables numéricas

```python
df.hist(bins=10,figsize=(20,10), grid = False, color = 'green')
plt.suptitle('Histogramas variables numéricas')
```

- Edad: Hay mayor concentración entre 18 y 25 años, lo que explica la primera barra más alta. Desde los 30 hasta los 60, la distribución se mantiene variada, pero sin extremos.
- IMC: El mayor número de personas está entre 25 y 35, categorizado como sobrepeso según la OMS.
- Hijos: La categoría 0 hijos es la más frecuente.
- Valor Seguro: El costo del seguro varía significativamente, pero la mayoría de personas paga tarifas relativamente bajas. Sin embargo, existen casos especiales con costos muy elevados.


## Gráfico 11. Relación de tres variables: valor seguro, IMC y el indicador de fumador

```python
sns.relplot(data=df, x="imc", y="valor_seguro", hue="fumador", palette = sns.color_palette("hls"))
plt.title('Relación entre IMC, Valor del Seguro y Estado Fumador')
plt.xlabel('IMC')
plt.ylabel('Valor Seguro')
```

Fumar es el factor que más incrementa el costo del seguro, pero el valor es más alto para los clientes con IMC entre 30 y 35.

## Gráfico 12. Valor seguro por sexo con indicador de fumador

```python
sns.boxplot(x=df["sexo"], y=df["valor_seguro"], hue = df["fumador"], palette = sns.color_palette("hls"))
plt.title('Valor seguro por sexo con indicador de fumador')
plt.xlabel('Sexo')
plt.ylabel('Valor Seguro')
```

Tanto hombres como mujeres fumadores presentan valores significativamente más altos que los no fumadores

## Gráfico 13. Conteo de clientes por region y sexo

```python
sns.countplot(data=df, x='region', hue='sexo', palette='hls')
plt.title('Conteo de Clientes por Región y Sexo')
plt.xlabel('Región')
plt.ylabel('Número de Clientes (Conteo)')
plt.show()
```

El número de clientes es similar entre hombres y mujeres en todas las regiones. Cundinamarca tiene el mayor conteo.
